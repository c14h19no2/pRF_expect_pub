```python
import os
import numpy as np
import platform
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from prf_expect.utils.viz import simple_colorbar
from prf_expect.utils.beh import calculate_visual_angle
from prf_expect.utils.io import load_settings
```

```python
settings = load_settings()
save_fig = False
```

```python
cwd = os.getcwd()
data_dir = Path(cwd).parent / "data"
figure_result_dir = Path(cwd).parent / "figures"

print('Running on {}'.format(platform.node()))
print('Current deriv folder is {}'.format(data_dir))
```

```python
fig_fn = figure_result_dir / 'polar_cbar.pdf'
fig =simple_colorbar(-np.pi, np.pi, "hsv", "horizontal", "Polar_hori")
plt.show()
```

```python
fig_fn = figure_result_dir / 'polar_cbar.pdf'
fig =simple_colorbar(-np.pi, np.pi, "hsv", "polar", "Polar")
plt.show()
if save_fig:
    fig.savefig(fig_fn, dpi=600, bbox_inches='tight', transparent=True)
```

```python
fig_fn = figure_result_dir / 'ecc_cbar.pdf'
rgb = np.loadtxt('eccentricity_cmap.cmap')
rgb = np.flip(rgb, axis=0)
cmap = mcolors.ListedColormap(rgb)
vis_angle = calculate_visual_angle(settings['monitor']["screen_size_cm"][1], settings['monitor']['screen_distance_cm'])/2*np.sqrt(2)
fig = simple_colorbar(0, vis_angle, cmap, "polar", "Ecc")
plt.show()
if save_fig:
    fig.savefig(fig_fn, dpi=600, bbox_inches='tight', transparent=True)
```

```python
fig_fn = figure_result_dir / 'paramb_cbar.pdf'
fig =simple_colorbar(-np.pi, np.pi, "viridis_r", "horizontal", "b")
fig.axes[0].set_xticks([])
fig.axes[0].set_xlabel('')
# remove borders
for ax in fig.axes:
    for spine in ax.spines.values():
        spine.set_visible(False)
plt.show()
if save_fig:
    fig.savefig(fig_fn, dpi=600, bbox_inches='tight', transparent=True)
```
